```python

import pandas as pd
import numpy as np
from pathlib import Path

bases_path = Path("bases")

checklist_A = bases_path / "HERING_FRANQUIA_checklist_RA_SMOOTH_0_2_2025-12-03.xlsx"
checklist_B = bases_path / "HERING_FRANQUIA_checklist_RA_SMOOTH_0_1_2025-12-02.xlsx"
df_check_antigo = pd.read_excel(checklist_B, usecols=["TRANSITO"], engine="calamine")
df_check_atual = pd.read_excel(checklist_A, usecols=["TRANSITO"], engine="calamine")

```

```python
total_antigo = df_check_antigo['TRANSITO'].sum()
total_atual = df_check_atual['TRANSITO'].sum()

# 2. Preven√ß√£o de erro: Se o antigo for 0, n√£o d√° para dividir (evita ZeroDivisionError)
if total_antigo == 0:
    print("Erro: O total antigo √© 0, imposs√≠vel calcular varia√ß√£o percentual.")
else:
    variacao = (total_atual - total_antigo) / total_antigo
    
    # Formatando para porcentagem (ex: -0.55 vira -55.0%)
    print(f"Total Antigo: {total_antigo:,.0f}")
    print(f"Total Atual:  {total_atual:,.0f}")
    print(f"Varia√ß√£o:     {variacao:.2%}")
    print("-" * 30)

   
    if variacao <= -0.5:
        print("üö® ALERTA CR√çTICO: Queda de tr√¢nsito superior a 50%! üö®")
        
    elif variacao < 0:
        print("Aten√ß√£o: Houve queda, mas dentro do aceit√°vel (< 50%).")
    else:
        print("Status: Tudo ok (Tr√¢nsito subiu ou manteve).")
```
