```python

import pandas as pd
import numpy as np
from pathlib import Path

bases_path = Path("bases")

checklist_A = bases_path / "HERING_FRANQUIA_checklist_RA_RA_FRANQUIA_2025-12-17.xlsx"
checklist_B = bases_path / "HERING_FRANQUIA_checklist_RA_NORMAL_RA_2025-12-10.xlsx"
df_check_antigo = pd.read_excel(checklist_B, usecols=["TRANSITO"], engine="calamine", sheet_name="CHECKLIST")
df_check_atual = pd.read_excel(checklist_A, usecols=["TRANSITO"], engine="calamine", sheet_name="CHECKLIST")

```

```python
total_antigo = df_check_antigo['TRANSITO'].sum()
total_atual = df_check_atual['TRANSITO'].sum()

# 2. Preven칞칚o de erro: Se o antigo for 0, n칚o d치 para dividir (evita ZeroDivisionError)
if total_antigo == 0:
    print("Erro: O total antigo 칠 0, imposs칤vel calcular varia칞칚o percentual.")
else:
    variacao = (total_atual - total_antigo) / total_antigo
    
    # Formatando para porcentagem (ex: -0.55 vira -55.0%)
    print(f"Total Antigo: {total_antigo:,.0f}")
    print(f"Total Atual:  {total_atual:,.0f}")
    print(f"Varia칞칚o:     {variacao:.2%}")
    print("-" * 30)

   
    if variacao <= -0.5:
        print("游뚿 ALERTA CR칈TICO: Queda de tr칙nsito superior a 50%! 游뚿")
        
    elif variacao < 0:
        print("Aten칞칚o: Houve queda, mas dentro do aceit치vel (< 50%).")
    else:
        print("Status: Tudo ok (Tr칙nsito subiu ou manteve).")
```
